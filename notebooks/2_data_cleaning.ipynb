```python
import pandas as pd

spot = pd.read_csv("../data/nifty_spot_5min.csv", parse_dates=["timestamp"])
futures = pd.read_csv("../data/nifty_futures_5min.csv", parse_dates=["timestamp"])

df = pd.merge_asof(
    spot.sort_values("timestamp"),
    futures.sort_values("timestamp"),
    on="timestamp",
    suffixes=("_spot", "_fut")
)

df.to_csv("../data/nifty_merged_5min.csv", index=False)
print("Merged data saved ✅")
```

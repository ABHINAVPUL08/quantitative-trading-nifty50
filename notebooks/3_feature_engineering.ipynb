```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv("../data/nifty_merged_5min.csv", parse_dates=["timestamp"])

df["ema_5"] = df["close_spot"].ewm(span=5).mean()
df["ema_15"] = df["close_spot"].ewm(span=15).mean()
df["returns"] = df["close_spot"].pct_change()

df.to_csv("../data/nifty_features_5min.csv", index=False)

plt.figure(figsize=(12,5))
plt.plot(df["timestamp"], df["close_spot"], label="Price")
plt.plot(df["timestamp"], df["ema_5"], label="EMA 5")
plt.plot(df["timestamp"], df["ema_15"], label="EMA 15")
plt.legend()
plt.title("EMA Feature Engineering")
plt.show()

print("Features created & plot shown ")
```
